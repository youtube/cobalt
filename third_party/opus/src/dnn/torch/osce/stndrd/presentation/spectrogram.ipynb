```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import wavfile
import scipy.signal
from playback import make_playback_animation

# coding noise demonstration
for br in ["inf", 24000, 15000, 12000, 9000, 8000, 7000, 6000]:
    fs, x = wavfile.read(f'data/a3_short_opus_{br}bps.wav')
    spec, freqs, t, im = plt.specgram(x, NFFT=512, cmap='inferno', noverlap=256 + 128, pad_to=4096)
    spec = 10*np.log10(spec)
    
    make_playback_animation(f'animations/opus_{br}bps.mp4', spec, len(x)/16)
    
```
